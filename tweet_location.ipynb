# Classifying Tweets

```python
import pandas as pd
pd.set_option('display.max_columns', 50)

new_york_tweets = pd.read_json("new_york.json", lines=True) #Tweets from New York
new_york_tweets.head()
```

```python
new_york_tweets.shape
```

```python
new_york_tweets.columns
```

```python
london_tweets = pd.read_json("london.json", lines=True) #Tweets from London
london_tweets.head()
```

```python
london_tweets.shape
```

```python
paris_tweets = pd.read_json("paris.json", lines=True) #Tweets from Paris
paris_tweets.head()
```

```python
paris_tweets.shape
```

# Classifying using language: Naive Bayes Classifier

```python
new_york_text = new_york_tweets["text"].tolist()
london_text = london_tweets["text"].tolist()
paris_text = paris_tweets["text"].tolist()
all_tweets = new_york_text + london_text + paris_text
labels = [0] * len(new_york_text) + [1] * len(london_text) + [2] * len(paris_text) #Creating labels {0:Newyork,1:London,2:Paris}
```

# Making a Training and Test Set


```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(all_tweets, labels, test_size=0.2, random_state=1)
print(len(X_train), len(X_test))
```

# Making the Count Vectors

```python
from sklearn.feature_extraction.text import CountVectorizer
counter = CountVectorizer().fit(X_train)
train_counts = counter.transform(X_train)
test_counts = counter.transform(X_test)
```

```python
print(X_train[3])
print(train_counts[3])
```

# Train and Test the Naive Bayes Classifier

```python
from sklearn.naive_bayes import MultinomialNB
classifier = MultinomialNB().fit(train_counts, y_train)
predictions = classifier.predict(test_counts)
```

# Model Evaluation



```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, predictions))
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, predictions)
```

# Conclusion

The classifier predicts tweets that were actually from New York as either New York tweets or London tweets, but almost never Paris tweets. Similarly, the classifier rarely misclassifies the tweets that were actually from Paris. Tweets coming from two English speaking countries are harder to distinguish than tweets in different languages.
